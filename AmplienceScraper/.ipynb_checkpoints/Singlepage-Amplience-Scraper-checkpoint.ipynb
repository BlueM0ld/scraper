Using beautiful soup 
bs2json converts the Html 

Instructions for this file : 

Two different methods to scrap the page 

1. HTML file to JSON file then formated
2. HTML file, format HTML file, Then convert to JSON file 

**First we scraped the page using requests & storing it in a BS object**

```python
import requests
from bs4 import BeautifulSoup
URL = "https://amplience.com/developers/docs/concepts/"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())

```

```python
#This step strips the html file and gives back only the text 

for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()

# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

#print(text)
```

```python
#This Step converts Text to Json 
import json

# convert into JSON:
y = json.dumps(text)

with open('json_data.json', 'w') as outfile:
    outfile.write(y)
    
# the result is a JSON string:
print(y)
```
